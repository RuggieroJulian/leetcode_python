```python
keyboard="abcdefghijklmnopqrstuvwxyz"
#getting a dictionary of time {'a':0, 'b':1, 'c':2 ..... 'z':26}
dictk={x:i for i,x in enumerate(keyboard)} 
text="cba"
position=0
time=0
#loop throw the letter in text
for letter in text:
    #add to time the abs between current position and the index of the letter that is being read (dictk[letter])
    time+=abs(position-dictk[letter])
    #store the letter read as the new position
    position=dictk[letter]
time
```
