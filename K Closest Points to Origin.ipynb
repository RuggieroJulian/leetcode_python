#### Leetcode Solution

```python
def kClosestLeetcode(points,k):
    import math
    listsqrt=[]
    #1st create a list of the euclidian distance
    for i,x in enumerate(points):
        listsqrt.append(math.sqrt((points[i][0]**2)+(points[i][1]**2)))
    listsqrt
    #use zip and sort to have [(1.0, [0, 1]), (1.0, [1, 0])]
    #remember sort applies to the first element, so will set the ones with the smallest euclidian distance first.
    #x for _,x means no matter what you have before the comma on the pair, I just want what is next, which is [0,1] and [1,0]
    return [x for _,x in sorted(zip(listsqrt,points))][:k]
kClosestLeetcode([[0,1],[1,0]],2)
```

#### Omar's Solution with Lambda Function

```python
def kClosestOmar(points, K: int):
    distance = sorted(points, key = lambda x: ((x[0]**2+ x[1]**2))**0.5)
    return distance[:K]
kClosestOmar([[0,1],[1,0]],2)
```
