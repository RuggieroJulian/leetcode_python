```python
import numpy as np
def largest_subarray(arrayA, number):
    a=arrayA
    k=number
    subarrays=[]
    for i in range(len(a)-1):
        if i+4<=len(a):
            subarrays.append(a[i:k+i])
        else:
            break
    #transforms the lists of lists to a numpy ndarray, to be able to compare columns        
    y=np.array([np.array(xi) for xi in subarrays])
    #the number of rows in the ndarray will be y.shape[0], so...
    row=0
    maxValue=0
    #iterating throw columns..
    for i in range(y.shape[1]-1):
        if row == y.shape[0]-1:
            #end of rows, leave for
            break
        else:
            #still have rows to compare
            row+=1
            if max(y[:,i]) == min(y[:,i]):
                #if the min and the max are equal, means all the column values are equal
                break
            else:
                #get the index value for the column that has the max value inside
                maxValue= np.where(y[:,i] == np.amax(y[:,i]))[0][0]
                break
    return subarrays[maxValue]
largest_subarray([1,4,3,2,5], 4)
```
