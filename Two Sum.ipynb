```python
#Time and space complexity

# Working ok, but O**2 solution
def twoSum(nums, target):
    listn=nums
    targetn=target
    leng=len(listn)
    result=[]
    for i in range(leng-1):
        for j in range(i+1,leng):
            if i < leng-1:
                #still have values ahead
                if listn[i]+listn[j]==targetn:
                    result.append(i)
                    result.append(j)
                    return result
```

```python
# Better solution, WORKING
def twoSumBetterWorking(nums, target):
    # two point
    nums_index = [(v, index) for index, v in enumerate(nums)]
    nums_index.sort()
    begin, end = 0, len(nums) - 1
    while begin < end:
        curr = nums_index[begin][0] + nums_index[end][0]
        if curr == target:
            return [nums_index[begin][1], nums_index[end][1]]
        elif curr < target:
            begin += 1
        else:
            end -= 1
```
