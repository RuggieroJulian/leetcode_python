### Level order traversal, Level by Level - BFS Iterative

Breadth-First Search is an algorithm to traverse or search in data structures like a tree or a graph. The algorithm starts with a root node and visit the node itself first. Then traverse its neighbors, traverse its second level neighbors, traverse its third level neighbors, so on and so forth.



Given a binary tree, return the level order traversal of its nodes' values. (ie, from left to right, level by level).

Given binary tree [3,9,20,null,null,15,7],


return its level order traversal as:


```python
# importing "collections" for deque operations 
import collections 

class Node(object):
    # init is the constructor
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
class BinaryTree(object):
    def __init__(self, root):
        self.root = Node(root)
        
    def levelOrderTraversal(self, root):
        """BFS using Queue and Collections Library"""
        if root is None:
            return []
        
        # initializing deque
        queue = collections.deque([]) 
        output=[]
        queue.append(root)
        countLeft=0
        countRight=0
        while (len(queue)!=0):
            # using popleft() to delete element from left end 
            root=queue.popleft()
            output.append(root.value)
            if (root.left is not None):
                queue.append(root.left)
                countLeft+=1
            if (root.right is not None):
                queue.append(root.right)
                countRight+=1
        # IF WANT TO FIND OUT THE NUMBER OF LEVELS OR Maximum Depth of Binary Tree
        # return max(countLeft,countRight)
        # IF WANT TO HAVE ELEMENTS IN ORDER CORRESPONDING TO THEIR LEVELS
        return output
```

```python
#       1
#    /    \
#   2      3
#  / \    / \
# 4  5   6  7
tree=BinaryTree(1)
tree.root.left=Node(2)
tree.root.right=Node(3)
tree.root.left.left=Node(4)
tree.root.left.right=Node(5)
tree.root.right.left=Node(6)
tree.root.right.right=Node(7)
```

```python
# Testing Level Order
tree.levelOrderTraversal(tree.root)
```
